```python
# Leitura da imagem
img = cv2.imread('teste.jpg')
# Conversão para escala de cinza
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Aplicação do filtro Gaussiano
suave = cv2.GaussianBlur(img, (7, 7), 0)
# Binarização
(T, bin) = cv2.threshold(suave, 0, 255, cv2.THRESH_BINARY)
(T, binI) = cv2.threshold(suave, 0, 255, cv2.THRESH_BINARY_INV)
# Combinação dos resultados
resultado = np.vstack([
    np.hstack([suave, bin]),
    np.hstack([binI, cv2.bitwise_and(img, img, mask=binI)])
])
# Exibição do resultado no Colab
cv2_imshow(resultado)

```
